# Introduction to Prompt Engineering (LiteLLM)
Prompt engineering is the process of designing and optimizing prompts for natural language processing tasks. It involves selecting the right prompts, tuning their parameters, and evaluating their performance. Prompt engineering is crucial for achieving high accuracy and efficiency in NLP models. In this section, we will explore the basics of prompt engineering using LiteLLM Proxy as the provider.

### Exercise 1: Tokenization
Explore Tokenization using tiktoken, an open-source fast tokenizer from OpenAI
See [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) for more examples.

````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
text = "Привет! Это пример текста на русском языке 😊 Посмотрим, как он разбивается на токены в модели.".
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### Exercise 2: Validate LiteLLM Proxy Setup

Run the code below to verify that your LiteLLM Proxy endpoint is set up correctly. The code just tries a simple basic prompt and validates the completion. Input `oh say can you see` should complete along the lines of `by the dawn's early light..`

````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
    base_url=os.environ["LITELLM_BASE_URL"],
    api_key=os.environ["LITELLM_API_KEY"],
)

deployment = os.environ.get("LITELLM_MODEL", "gpt-4o")

def get_completion(prompt):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=deployment,
        messages=messages,
        temperature=0,
        max_tokens=1024
    )
    return response.choices[0].message.content

text = """
Artificial intelligence can help summarize articles and answer questions.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

````python
# Chain-of-thought example

prompt = """
Maria has 24 apples. She gives 7 apples to her friend and then buys 15 more apples.
Reason step by step and tell me how many apples Maria has now.
"""

response = get_completion(prompt)
print(response)
````

````python
# Generated knowledge

prompt = """
Facts:
- Renewable energy comes from natural sources like sunlight and wind.
- It produces little to no greenhouse gas emissions.
- It reduces dependence on fossil fuels.

Using the facts above, explain why renewable energy is important.
"""

response = get_completion(prompt)
print(response)
````

````python
# Least-to-Most example
prompt = """
Break down the problem and solve it step by step.

Problem:
A student scored 70, 80, and 90 on three tests. What is the average score?

Steps:
1. Find the total score.
2. Count the number of tests.
3. Calculate the average.

Final answer:
"""

response = get_completion(prompt)
print(response)
````

````python
# Self-refine example

# Step 1: Initial answer
initial_prompt = """
Explain why exercise is important.
"""

initial_response = get_completion(initial_prompt)
print("Initial response:")
print(initial_response)


# Step 2: Your critique
critique = """
The answer is too general. Add specific examples and mention both physical and mental benefits.
"""

# Step 3: Refine based on critique
refine_prompt = f"""
Improve the following answer based on this feedback:

Answer:
{initial_response}

Feedback:
{critique}
"""

refined_response = get_completion(refine_prompt)
print("\nRefined response:")
print(refined_response)
````

````python
# Maieutic prompting example

# Step 1: Initial answer
initial_prompt = """
What is the capital of France and why is it important?
"""

initial_response = get_completion(initial_prompt)
print("Initial response:")
print(initial_response)


# Step 2: Ask follow-up questions to verify and deepen understanding
followup_prompt = f"""
Based on the previous answer:

{initial_response}

1. Explain why the capital city plays an important role in a country.
2. What historical factors made this city the capital?
3. Are there any alternative cities that could have been the capital? Why or why not?
"""

followup_response = get_completion(followup_prompt)
print("\nFollow-up explanation:")
print(followup_response)
````
