### Build your first prompt  
This short exercise will provide a basic introduction for submitting prompts to an LLM via LiteLLM Proxy for a simple task "summarization".


**Steps**:  
1. Install OpenAI library in your python environment  
2. Load standard helper libraries and set your LiteLLM Proxy credentials  
3. Choose a model for your task  
4. Create a simple prompt for the model  
5. Submit your request to the model API!

### 1. Install OpenAI

```python
%pip install openai python-dotenv
```

### 2. Import helper libraries and instantiate credentials

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

# LiteLLM Proxy provides an OpenAI-compatible API
client = OpenAI(
    base_url=os.environ["LITELLM_BASE_URL"],
    api_key=os.environ["LITELLM_API_KEY"],
)
```

### 3. Finding the right model  
The model name depends on what is configured in your LiteLLM Proxy. Common choices include GPT-4o, Claude, Llama, Mistral, and others.

```python
# Select the model configured in your LiteLLM Proxy
model = os.environ.get("LITELLM_MODEL", "gpt-4o")
```

## 4. Prompt Design  

"The magic of large language models is that by being trained to minimize this prediction error over vast quantities of text, the models end up learning concepts useful for these predictions. For example, they learn concepts like"(1):

* how to spell
* how grammar works
* how to paraphrase
* how to answer questions
* how to hold a conversation
* how to write in many languages
* how to code
* etc.

#### There are three basic guidelines to creating prompts:

**Show and tell**. Make it clear what you want either through instructions, examples, or a combination of the two.

**Provide quality data**. If you're trying to build a classifier or get the model to follow a pattern, make sure that there are enough examples.

**Check your settings.** The temperature and top_p settings control how deterministic the model is in generating a response.

### 5. Submit!

```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.chat.completions.create(
  model=model,
  messages = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},])

response.choices[0].message.content
```
