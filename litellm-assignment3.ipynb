## Exercise - your first text generation app

Now that we learned how to set up LiteLLM Proxy, it's time to build your first text generation app.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

# LiteLLM Proxy provides an OpenAI-compatible API
client = OpenAI(
    base_url=os.environ["LITELLM_BASE_URL"],
    api_key=os.environ["LITELLM_API_KEY"],
)

deployment = os.environ.get("LITELLM_MODEL", "gpt-4o")

# add your completion code
prompt = "Complete the following: There are no rules and instructions in the IT company"
messages = [{"role": "user", "content": prompt}]

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
```

## Exercise - build a recipe generator

Let's build a recipe generator that takes ingredients and returns recipes.

```python
prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
```

### Improve by adding filter and shopping list

Let's make the code flexible with user inputs for ingredients, number of recipes, and filters.

```python
no_recipes = 2
ingredients = "chicken, potatoes, carrots"
diet = "none"

prompt = f"""
Generate {no_recipes} simple recipes using these ingredients: {ingredients}.
Diet restriction: {diet}.

For each recipe include only:
- Recipe name
- Extra ingredients needed
- 3 short cooking steps
"""

messages = [{"role": "user", "content": prompt}]

completion = client.chat.completions.create(
    model=deployment,
    messages=messages,
    max_tokens=300,
    temperature=0.1,
    timeout=30
)

recipes = completion.choices[0].message.content
print("Recipes:")
print(recipes)

prompt_shopping = f"""
Ingredients already at home: {ingredients}

Based on these recipes:
{recipes}

Create one combined shopping list.
Only include ingredients that are NOT already at home.
Keep it short.
"""

messages = [{"role": "user", "content": prompt_shopping}]

completion = client.chat.completions.create(
    model=deployment,
    messages=messages,
    max_tokens=150,
    temperature=0.1,
    timeout=30
)

print("\n===== Shopping list =====\n")
print(completion.choices[0].message.content)
```

```python
# Study Buddy app

topic = "Python"
question = input("Ask your study question: ")

prompt = f"""
You are a helpful study buddy for the topic: {topic}.

Your job is to help the student learn.
Answer the student's question clearly and simply.

Rules:
- Explain in beginner-friendly language
- Give a short and accurate answer
- If helpful, include a simple example
- If the question is about code, include a short code example
- End with one short study tip or follow-up question

Student question:
{question}
"""

messages = [{"role": "user", "content": prompt}]

completion = client.chat.completions.create(
    model=deployment,
    messages=messages,
    max_tokens=500,
    temperature=0.3,
    timeout=30
)

print("Study Buddy answer:\n")
print(completion.choices[0].message.content)
```

```python
# History Bot

character = input("Choose a historical character: ")
question = input("Ask your question: ")

prompt = f"""
You are a history bot roleplaying as {character}.

Answer in first person, as if you are this historical character speaking about your life and times.

Rules:
- Stay historically grounded
- Be engaging and vivid
- Keep the answer clear and easy to understand
- If something is uncertain or debated, say so
- Mention important historical context when relevant
- Keep the answer under 250 words

Question:
{question}
"""

messages = [{"role": "user", "content": prompt}]

completion = client.chat.completions.create(
    model=deployment,
    messages=messages,
    max_tokens=500,
    temperature=0.7,
    timeout=30
)

print("\nHistory Bot answer:\n")
print(completion.choices[0].message.content)
```
